```python
from datetime import date
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
def wrangle(tickers):
    today = date.today()
    end_date = pd.to_datetime(today)
    start_date = end_date - pd.DateOffset(252 * 10)
    data = {}
    for ticker in tickers:
        df = yf.download(ticker, start=start_date, end=end_date)
        df.index.name = "date"
        df.columns = df.columns.str.lower()
        data[ticker] = df
    return data
```

```python
def plot_stocks(tickers, data):
    fig,ax = plt.subplots(figsize=(15,6), dpi=200)
    for ticker in tickers:
        data[ticker]["adj close"].plot(ax=ax, label=ticker)
    plt.xlabel("Date")
    plt.ylabel("Close")
    plt.title("Stocks Adjust Close Graph")
    plt.legend()
    return fig
```

```python
def perc_calc(tickers, data, start_date, end_date):
    perc_data = {}
    for ticker in tickers:
        if start_date not in data[ticker]["adj close"].index:
            print(f"This Start Date is not available in {ticker} data.")
        if end_date not in data[ticker]["adj close"].index:
            print(f"This End Date is not available in {ticker} data.")

        adj_close_start = data[ticker]["adj close"][start_date]
        adj_close_end = data[ticker]["adj close"][end_date]
        perc = 100*(adj_close_end - adj_close_start)/adj_close_start
        perc_data[ticker] = f"Percent Change: {np.round(perc, 2)}"
    return perc_data
```

```python
def invested_shares_plot(data, tickers, amount, start_date, end_date):
    fig,ax = plt.subplots(figsize=(15,6), dpi=200)
    invest_data = []
    for index , df in data.items():
        df = df[df.index >= pd.to_datetime(start_date)]
        df = df[df.index <= pd.to_datetime(end_date)]
        ret = df["adj close"].pct_change(1).dropna()
        cum_ret = (ret+1).cumprod()
        result = cum_ret * amount
        data = {
            "amount" : amount,
            "ticker" : index,
            "start_date" : pd.to_datetime(start_date).isoformat(),
            "end_date" : pd.to_datetime(end_date).isoformat(),
            "result" : np.round(result.iloc[-1], 2),
            "shares" : np.round(amount/df["adj close"].iloc[-1]),
            "shape_ratio" : compute_sharp_ratio(df),
            "message" : f"{amount}$ Amount Investment in {index} will be resulted: {np.round(result.iloc[-1], 2)}"
        } 
        invest_data.append(data)
        result.plot(ax=ax, label=index)
    plt.xlabel("Date")
    plt.ylabel("Result")
    plt.title(f"{amount}$ Investment in {tickers} Result Graph")
    plt.legend()
    return invest_data
```

```python
def compute_sharp_ratio(data, risk_free_rate=0):
    mean = data["adj close"].pct_change(1).mean()
    std = data["adj close"].pct_change(1).std()
    sharpe_ratio = (mean - risk_free_rate) / std
    return np.round(sharpe_ratio,4)
```

```python
tickers = ["AAPL", "MSFT"]
data = wrangle(tickers=tickers)
data["AAPL"].tail()
```

```python
fig = plot_stocks(tickers,data)
fig.show()
```
